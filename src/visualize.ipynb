```python
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('training_results.pkl', 'rb') as f:
    train_losses, eval_losses, train_corrects, eval_corrects, best_accuracy = pickle.load(f)

train_accuracies = [correct / int(11992*0.8) for correct in train_corrects]
eval_accuracies = [correct / int(11992*0.2) for correct in eval_corrects]
```

```python
plt.figure(figsize=(10,6))
plt.plot(train_losses, label="Training Loss", color='blue')
plt.plot(eval_losses, label="Validation Loss", color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title("Training and Validation Loss over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_accuracies, label='Training Accuracy', marker='o', color='blue')
plt.plot(eval_accuracies, label='Validation Accuracy', marker='o', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
